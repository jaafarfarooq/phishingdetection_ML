# Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python

y = np.asarray(df[['Result']])
b=(list(df))
#print(b[:-1])
x = np.asarray(df[b[:-1]])
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(X_train,y_train)
```

```python
y_pred=dtree.predict(X_test)
y_trainpred=dtree.predict(X_train)
```

### Confusion Matrix and Accuracy for Decision tree

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm_dtree=confusion_matrix(y_test,y_pred)
print("True Possitives: ", cm_dtree[0][0],"True Negatives: ",cm_dtree[0][1])
print("False Possitives: ", cm_dtree[1][0],"False Negatives: ",cm_dtree[1][1])
accuracy_dtree=accuracy_score(y_test,y_pred)
print("Decision Tree Accuracy: ",accuracy_dtree)
```

### Classification Report

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
